```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import datetime
```

```python
# Handle date time conversions between pandas and matplotlib
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

# Use white grid plot background from seaborn
sns.set(font_scale=1.5, style="whitegrid")
```

```python
#CSV to Load
zillow_data= 'zillow_new.csv'
zillow_df=pd.read_csv(zillow_data, na_values=['999.99'])
zillow_df.head()
```

```python
zillow_df[['City', "State"]]=zillow_df.MsaName.str.split(",", expand=True,)
zillow_df.rename(columns={'RegionName': 'ZIP'}, inplace=True)
zillow_df = zillow_df.drop(['MsaName','RegionID', 'Unnamed: 0', 'SizeRank'], axis=1)
zillow_df.head()
```

```python
zillow_dft = pd.DataFrame.transpose(zillow_df)
zillow_dft.head()
```

```python
zillow_dft.columns = zillow_dft.iloc[0]
zillow_dft.drop(index=zillow_dft.index[0], 
        axis=0, 
        inplace=True)
zillow_dft.tail()
```

```python
zillow_dft.to_csv('zillow_date_column.csv')
```
